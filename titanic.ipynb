```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Загрузка данных
titanic_df = pd.read_csv('/kaggle/input/titanicex/Titanic.csv')
titanic_df.head()
```

```python
# 1. График распределения выживаемости по классам
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)
plt.title('Выживаемость в зависимости от класса')
plt.xlabel('Класс билета')
plt.ylabel('Количество')
plt.legend(title='Выживание', labels=['Не выжил', 'Выжил'])
plt.show()
```

вывод: Чем выше класс тем больше шанс выжить

```python
# 2. График распределения выживаемости по полу
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.title('Выживаемость в зависимости от пола')
plt.xlabel('Пол')
plt.ylabel('Количество')
plt.legend(title='Выживание', labels=['Не выжил', 'Выжил'])
plt.show()
```

Вывод: Женщин выжило больше

```python
# 3. Распределение возраста среди выживших и погибших
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_df, x='Age', hue='Survived', multiple="stack", kde=True)
plt.title('Распределение возраста среди выживших и погибших')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.legend(title='Выживание', labels=['Не выжил', 'Выжил'])
plt.show()
```

```python
# 4. Выживаемость в зависимости от количества родственников на борту
# Создадим новый столбец, суммирующий количество братьев/сестер и родителей/детей
titanic_df['RelativesOnBoard'] = titanic_df['SibSp'] + titanic_df['Parch']

plt.figure(figsize=(10, 6))
sns.countplot(x='RelativesOnBoard', hue='Survived', data=titanic_df)
plt.title('Выживаемость в зависимости от количества родственников')
plt.xlabel('Количество родственников на борту')
plt.ylabel('Количество')
plt.legend(title='Выживание', labels=['Не выжил', 'Выжил'])
plt.show()
```

Вывод: те у кого не было родственников на борту чаще не выживали, у кого было до 3х родственников выживали чаще

```python
# 5. Распределение стоимости билетов среди выживших и погибших
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_df, x='Fare', hue='Survived', kde=True)
plt.title('Распределение стоимости билетов среди выживших и погибших')
plt.xlabel('Стоимость билета')
plt.ylabel('Количество')
plt.legend(title='Выживание', labels=['Не выжил', 'Выжил'])
plt.show()
```

Вывод: Люди с более дорогим билетов выживали больше, что логично

```python
# 6. Выживаемость в зависимости от порта отправления
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='Survived', data=titanic_df)
plt.title('Выживаемость в зависимости от порта отправления')
plt.xlabel('Порт отправления')
plt.ylabel('Количество')
plt.legend(title='Выживание', labels=['Не выжил', 'Выжил'])
plt.show()
```

Вывод: у кого был порт отправления S погибли в большем колличестве как и порт Q, тогда как порт C выжило больше(но скорее всего просто так совпало) 

```python
# 7. Процент выживших в зависимости от пола и класса
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=titanic_df, ci=None)
plt.title('Процент выживших в зависимости от пола и класса')
plt.xlabel('Класс')
plt.ylabel('Процент выживших')
plt.show()
```

Вывод: Женщины выживали чаще чем мужчины вне зависимости от класса
